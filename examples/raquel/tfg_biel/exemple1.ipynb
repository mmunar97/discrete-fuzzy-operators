```python
from discrete_fuzzy_operators.base.operators.binary_operators.unit.suboperators.fuzzy_unit_aggregation_suboperators.unit_tnorm import FuzzyUnitTnorm
import numpy as np
points = 200
```

```python
a=0.6
def TA(x,y):
    return np.minimum(x,y)

def t(x,y):
    if x == 0 or y == 0:
        return 0
    elif x == 1:
        return y
    elif y == 1:
        return x
    elif not (x-a)**2 + (y-a)**2 >= a**2 + (a-1)**2:
        return TA(x,y)
    else:
        return 0
T = FuzzyUnitTnorm(t)
T.contour_plot_operator(points,points)
```
